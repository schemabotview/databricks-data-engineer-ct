## Joins — the seven types

Spark supports the full SQL join family plus two Spark-specific shapes (`semi`, `anti`). Memorise this table — the exam quotes it almost verbatim:

| Join type | Returns |
|---|---|
| `inner` | Rows where the key matches on both sides |
| `left` (`left_outer`) | All left rows; nulls on right when no match |
| `right` (`right_outer`) | All right rows; nulls on left when no match |
| `full` (`full_outer`, `outer`) | All rows from both sides; nulls on the missing side |
| `left_semi` | Left rows that **have a match** on the right — *no columns from right* |
| `left_anti` | Left rows that **do NOT match** on the right |
| `cross` | Cartesian product — rarely the right answer |

Three API shapes for the join condition:

```python
# Same key column on both sides — pass a string
df.join(customers, on="customer_id", how="inner")

# Different names — pass a column expression
df.join(customers, df.cust_id == customers.customer_id, how="left")

# Multi-key — pass a list of column names
df.join(other, on=["customer_id", "account_id"], how="inner")
```

**`left_semi` is the exam favourite** for "give me the customers who have at least one transaction." A plain `select` would scan every matching row; `left_semi` short-circuits at the first match per key and brings back *only* the left columns. Its mirror, `left_anti`, answers "customers with *no* transactions" — the standard anti-join for finding orphans.
