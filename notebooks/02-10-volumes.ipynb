## Volumes — governed storage for files, not rows

Tables are for rows. **Volumes** are UC's answer for everything that isn't tabular — raw JSON dumps, PDFs of loan agreements, model artifacts, libraries, image datasets.

A volume is a UC object like any other: `catalog.schema.volume_name`. From a notebook you read and write its files under the path `/Volumes/catalog/schema/volume_name/...`.

Two flavours, mirroring tables:

- **Managed volume** — UC owns the storage location. Use it for unstructured data your team produces.
- **External volume** — UC catalogs files at a path you already own. Use it for data dropped in by external systems — the bank's SFTP landing zone for daily card extracts.

```sql
CREATE VOLUME fintech_dev.bronze.landing_zone
  COMMENT 'Daily card extracts dropped by the source system';
-- then in Python:  dbutils.fs.ls("/Volumes/fintech_dev/bronze/landing_zone")
```

Volumes **replace the legacy `dbfs:/mnt/...` mounts** — mounts had no UC permission model, volumes do. **On the exam, volumes are the answer** whenever the question involves governing access to files that aren't already Delta tables.
