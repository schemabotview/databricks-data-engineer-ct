## Unions & combining — `union`, `unionAll`, `unionByName`

Three operations that sound alike and behave differently. The exam tests the difference directly.

- **`union(other)`** — concatenates rows, matching columns by **position**. Both DataFrames must have the same number of columns in the same order. It does **NOT** deduplicate.
- **`unionAll(other)`** — a deprecated *alias* for `union`. Identical behaviour, older name.
- **`unionByName(other, allowMissingColumns=False)`** — matches columns by **name**, not position. Safer in production. With `allowMissingColumns=True`, columns missing on one side are filled with `NULL`.

```python
# Positional — column order must line up
all_txns = jan.union(feb)

# Name-based, tolerant of schema drift — prefer this
all_txns = jan.unionByName(feb, allowMissingColumns=True)
```

**The exam trap:** developers expect `union` to deduplicate, because SQL `UNION` does. Spark's `union` is closer to SQL **`UNION ALL`** — it keeps duplicates. To deduplicate, chain `.distinct()`:

```python
all_txns = jan.union(feb).distinct()
```

**Rule of thumb:** prefer `unionByName` in any code that has to survive schema changes — matching by position silently corrupts data the moment someone reorders a column.
