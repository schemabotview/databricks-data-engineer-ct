## Other inbound paths — partner connectors, JDBC, REST

Beyond managed connectors, three more paths cover the long tail.

**Partner connectors** — Fivetran, Airbyte, and friends, surfaced under *Partner Connect* with billing through Databricks. Reach for one when the source is in the long tail (Marketo, Stripe, Zendesk) that Lakeflow Connect doesn't cover yet, or you're already standardised on a partner tool. The exam only expects you to treat them as a *peer* option and prefer **managed** when the source is supported.

**JDBC / ODBC** — read an operational DB directly when there's no managed connector or you need a one-off pull:

```python
loans_df = (spark.read.format("jdbc")
    .option("url", "jdbc:postgresql://loans-prod:5432/loans")
    .option("dbtable", "public.loan_accounts")
    .option("user", dbutils.secrets.get("loans", "username"))
    .option("password", dbutils.secrets.get("loans", "password"))
    .load())
```

Credentials come from a **secret scope**, never hardcoded; parallelise big pulls with `partitionColumn` / `lowerBound` / `upperBound` / `numPartitions`; it's a **pull** pattern, usually wrapped in a nightly Lakeflow Jobs task.

**REST API** — for sources with neither a connector nor a driver: paginate, auth from secrets, and **land raw JSON to a UC volume first**, *then* ingest with Auto Loader / `COPY INTO`. Treat REST as a **landing strategy**, not a target.
