## The decision sheet — picking the right access control

The whole module on one page. Read the scenario, pick the control:

| Scenario | Pick |
|---|---|
| Give analysts read access to gold tables | `GRANT USE CATALOG` + `USE SCHEMA` + `SELECT` |
| Block one table for a group despite a catalog-level grant | **`DENY`** (beats GRANT) |
| Hide PAN from analysts but show last-4 to the fraud team | **Column mask** |
| Restrict compliance analysts to their region's rows | **Row filter** |
| Apply masks / filters to every PII-tagged column across hundreds of tables | **Unity Catalog ABAC** |
| Fine-grained control for a non-UC client, or a very simple case | **Dynamic view** (legacy) |
| Forensic audit: who read this table last Tuesday? | **`system.access.audit`** |
| Track column-level read/write lineage | **`system.access.table_lineage`** + Lineage tab |
| Production grants for a team | to a **group**, not individuals |

**Three tells that unlock most questions:**

- *"read access"* → remember the **`USE CATALOG` + `USE SCHEMA` + `SELECT`** trio (the omission trap).
- *"which columns' values"* → **column mask**; *"which rows"* → **row filter**; *"across many tables / centralised"* → **ABAC**.
- *"who accessed / when / lineage"* → **`system.access`**.

That completes the exam's seven sections. The recurring instinct that wins: when two answers compete, pick the more **managed, declarative, base-table-enforced, Databricks-recommended** one — here, that's masks / filters / ABAC over dynamic views, and groups over individual users.
