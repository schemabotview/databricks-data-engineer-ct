## Monitoring, repair & rerun

The Lakeflow Jobs UI is the exam's go-to for monitoring questions. Four things to know.

**Run history view** — every run of a job, with state, duration, and a **sparkline of durations over time**. Used to spot *trends* — "the job takes 2× longer than three months ago."

**Task graph view** — the DAG for a single run, coloured by state: **red = failed, grey = skipped, green = succeeded**. Used to find the **upstream blocker** — if `ingest_cards` is red and everything downstream is grey, the root cause is `ingest_cards`.

**Repair run** — rerun **only the failed tasks** of a previous run, with the same parameters, in the same DAG. Cheaper and safer than re-running the whole job.

**Run dashboard / Jobs metrics** — aggregate views (success rate, p95 duration) for capacity and SLA tracking.

**The exam's monitoring pattern:** a job's median duration tripled overnight — *where do you look first?* → the **run history view** for the trend, then the **task graph view** of a slow run to find the offending task. From there, module 08 goes deeper into Spark-level diagnostics.

**Why repair beats rerun:** re-running a six-hour job because the last task failed wastes five successful tasks. **Repair run** re-executes just the failed leaf, reusing the upstream results.
