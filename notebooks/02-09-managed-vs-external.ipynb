## Managed vs. external tables — and converting between them

Every UC table is one of two flavours, and the choice decides a lot downstream.

**Managed table** — Unity Catalog owns the metadata **and** the files. The files live in a UC-managed storage location tied to the catalog or schema. `DROP TABLE` removes the metadata and the files are deleted after retention.

**External table** — UC owns the metadata; the files live at a path **you** specify inside a UC external location. `DROP TABLE` removes only the metadata — the files stay.

| Question | Managed | External |
|---|---|---|
| Who owns the files? | Unity Catalog | You |
| Where do files live? | UC-managed location | A path you specify |
| `DROP TABLE`? | Metadata + files (after retention) | Metadata only; files preserved |
| Predictive Optimization? | Yes | No |
| Share files with non-Databricks tools? | No | Yes |

**The default the exam rewards: prefer managed.** External exists for specific reasons — a multi-tool data lake, files produced outside Databricks, a regulator requiring a fixed path you control. If none apply, choose managed.

You can convert either direction with `ALTER TABLE ... SET MANAGED` / `SET EXTERNAL`. Both are **metadata-level on the UC side** — the three-part name stays the same; physical files may or may not move depending on direction; and you need privileges on both the table and the target storage location.
