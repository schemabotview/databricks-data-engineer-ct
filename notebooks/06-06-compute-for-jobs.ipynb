## Compute for jobs — per-task choice

Tasks within one job can use **different compute**. Four choices, and when each wins:

- **Job cluster** — created at task start, torn down at task end. **Cheapest DBU rate; the default for production ETL.**
- **Serverless compute (jobs)** — second-scale startup, no cluster management. Best for **short tasks** where a 3-minute classic startup is wasted overhead.
- **All-purpose cluster** — long-lived, shared. **Avoid for scheduled jobs** — higher DBU rate and cross-job interference risk.
- **SQL warehouse** — **required** for SQL and Dashboard tasks; serverless SQL warehouse is the modern default.

**Compute mix in the bank's nightly job:**

- `ingest_cards` (Auto Loader notebook) → job cluster, DBR 14.x LTS.
- `silver_build` (heavy SQL notebook) → the *same* job cluster reused.
- `refresh_dashboard` (Dashboard task) → serverless SQL warehouse.
- `notify` (tiny notebook) → serverless jobs compute.

**Cluster reuse across tasks** — set on the cluster definition so several notebook tasks share one cluster and pay startup **once**. It saves real money on a chain of small steps.

**The exam's angle:** for scheduled production ETL, pick a **job cluster** (or serverless), never an all-purpose cluster. All-purpose is for interactive development; using it for a nightly job is the wrong (and pricier) answer.
