## Task types — the four the exam tests by name

Lakeflow Jobs supports a long list of task types. Four are asked about directly — know what each runs and where it runs:

| Task type | Runs | Compute |
|---|---|---|
| **Notebook** | A `.ipynb` from Workspace or a Git Folder | Cluster (job / all-purpose / serverless) |
| **SQL** | A saved query, alert, file, or dashboard | **SQL warehouse** |
| **Pipeline** | A Lakeflow Spark Declarative Pipeline (ex-DLT) | Pipeline compute (managed by the pipeline) |
| **Dashboard** | Refresh an AI/BI dashboard | **SQL warehouse** |

The **SQL task has four sub-variants**, mirroring DBSQL's four artifact types — query, alert, file, dashboard; *SQL task — query* is the most common.

**Secondary task types to recognise:**

- **JAR**, **Python script**, **Python wheel** — code packaged outside notebooks (the Automation Bundle workflow, module 07).
- **dbt** — run a dbt project task.
- **Run job** — run *another* Lakeflow Job (nested composition).
- **`for_each`** — loop a child task over an iterable.
- **`if/else` condition** — branch the DAG.

**The exam's angle:** match artifact to task type. Refreshing a dashboard → **Dashboard task**, not a Notebook task that happens to run SQL. Running a pipeline → **Pipeline task**, so the pipeline manages its own compute — you don't attach a cluster to it.
