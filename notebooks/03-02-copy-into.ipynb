## `COPY INTO` — idempotent file ingestion in pure SQL

`COPY INTO` is a SQL statement that loads files from cloud storage into a Delta table, **remembers which files it has already loaded**, and skips them on the next run. That one property — idempotency — makes it the right answer for daily-batch file loads.

```sql
COPY INTO fintech_dev.bronze.bank_accounts_raw
FROM '/Volumes/fintech_dev/bronze/landing_zone/accounts/'
FILEFORMAT = CSV
FORMAT_OPTIONS ('header' = 'true', 'inferSchema' = 'true')
COPY_OPTIONS  ('mergeSchema' = 'true');
```

- The target is a UC-governed Delta table; the source path is any object store the workspace can reach, usually via a UC **volume** or external location.
- After the first run, Delta records the loaded file paths — re-running is a **no-op** for already-loaded files.
- `FORMAT_OPTIONS` are the reader (source-format) options; `COPY_OPTIONS` are the Delta writer (target) options.

**Pick `COPY INTO` when:** a small-to-moderate number of files per run, a predictable cadence (a scheduled Lakeflow Jobs task), you want declarative SQL with no streaming machinery, and the format is supported (CSV, JSON, Avro, ORC, Parquet, XML, BINARYFILE).
