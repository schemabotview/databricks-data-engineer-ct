## AQE — what adaptive query execution fixes automatically

**Adaptive Query Execution (AQE)** re-optimises the physical plan **at runtime**, after each shuffle, using *actual* partition statistics instead of the planner's estimates. Three things it does for you:

- **Dynamic partition coalescing** — merges undersized post-shuffle partitions into fewer, larger, more balanced units. Fewer task-startup overheads.
- **Dynamic join-strategy switching** — promotes a sort-merge join to a **broadcast join** when one side turns out smaller than estimated at runtime.
- **Skew-join handling** — detects oversized partitions and splits them, so one slow task doesn't block the whole stage.

**AQE is on by default since Spark 3.2 / DBR 11.0** — on a modern Databricks workspace you have it for free. That's why the exam's correct answer to most performance questions is *confirm AQE (or its skew / coalesce handling) is enabled* — **not** manual tuning.

**Config keys to recognise:**

- `spark.sql.adaptive.enabled` — the master switch.
- `spark.sql.adaptive.coalescePartitions.enabled`
- `spark.sql.adaptive.skewJoin.enabled`
- `spark.sql.adaptive.localShuffleReader.enabled`

**The through-line of this module:** skew, shuffle-partition sizing, and join-strategy choice are the three big performance problems — and AQE addresses all three at runtime. Manual tuning is the fallback for when AQE has been ruled out; AQE is the default answer.
