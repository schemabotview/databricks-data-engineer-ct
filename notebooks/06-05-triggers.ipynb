## Triggers — time-based vs. data-driven (the exam decision)

A trigger is *what fires a job run*. Five options; the first three matter most:

- **Scheduled (cron / time-based)** — run at a quartz-cron time (`0 0 2 * * ?` = daily 02:00 UTC). Best when the source lands on a **predictable cadence**.
- **File arrival** — run when new files land in a path; Lakeflow Jobs polls and fires when it sees them. Best when landing time is **irregular**.
- **Table update** — run when a *Delta table* gets a new commit. Best when the dependency is **table state** — rebuild gold whenever `silver` changes.
- **Continuous** — run forever; only for a declarative pipeline in continuous mode.
- **Manual** — no trigger; a human or API starts it. Good for ad-hoc backfills.

**The exam frames it as one tradeoff — time-based vs. data-driven:**

- **Time-based (cron)** when data lands predictably, you want a stable window for SLAs and cost forecasting, and "fresh as of midnight" is good enough.
- **Data-driven (file arrival / table update)** when landing time is variable, you want to **minimise latency**, and empty polled runs would be wasteful.

**The classic scenario:** a partner drops a file at an unpredictable time each night; ingest as soon as it lands, without an empty hourly job. → **File arrival**. Cron is the distractor that wastes compute.
