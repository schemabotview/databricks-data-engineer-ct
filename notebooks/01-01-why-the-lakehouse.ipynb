## Why the lakehouse — warehouse vs. lake, and why they fused

For about twenty years, analytics teams had to choose between two very different systems, and neither one could do the whole job.

A **data warehouse** stores structured, tabular data and is built for fast SQL and business intelligence. It gives you ACID transactions, strong governance, and predictable query performance. The price you pay: a rigid schema you must define up front, proprietary storage you can't read with other tools, and a cost that climbs steeply as data grows. Warehouses were never designed for unstructured data, streaming, or machine learning.

A **data lake** is the opposite. It's cheap object storage — Amazon S3, Azure Data Lake Storage, Google Cloud Storage — holding files in any format: JSON, CSV, Parquet, images, logs. It's open and scales endlessly, which is why data science and streaming gravitated to it. But a raw lake has no transactions, no schema enforcement, and no governance. Concurrent writes corrupt tables, a failed job leaves half-written files behind, and with no rules the lake slowly rots into a *data swamp* nobody trusts.

So most companies ran **both**: land everything in a lake, then copy the curated slice into a warehouse for BI. That split is the real problem. Data is duplicated, the two copies drift out of sync, governance is spread across two tools, and you pay for two platforms plus all the pipelines that shuttle data between them.

The **lakehouse** collapses the two into one. You keep the lake's cheap, open object storage, but add a transaction and metadata layer on top — so the *same* files gain ACID guarantees, schema enforcement, and governance. Now BI, streaming, and machine learning all read one governed copy.

On Databricks that layer is **Delta Lake** (transactions on the files) plus **Unity Catalog** (governance over them), and the product built around it is the **Data Intelligence Platform** — the subject of this module.
