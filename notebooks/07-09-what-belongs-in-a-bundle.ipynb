## What belongs in a bundle, what doesn't

A bundle is for **workspace-shaped code**, not for data or infrastructure.

| Asset | In a bundle? |
|---|---|
| Lakeflow Jobs | ✅ |
| Lakeflow Spark Declarative Pipelines | ✅ |
| Notebooks invoked by the above | ✅ |
| Python source files / wheels | ✅ |
| ML experiments + models | ✅ |
| Volumes + schemas the bundle creates | ✅ |
| AI/BI dashboards | ✅ |
| Raw data files | ❌ — bundles aren't a data deployment tool |
| Unity Catalog catalogs themselves | ❌ — created out of band, then referenced |
| Service principals + IAM roles | ❌ — managed by your cloud / identity provider |
| Secret *values* | ❌ — only secret *scope references* go in a bundle |

**The rule:** if the asset is **workspace-shaped code**, put it in a bundle. If it's **infrastructure** (catalogs, principals, network) or **data**, keep it in Terraform / cloud IaC / your data pipelines.

**Why the boundary matters:** bundles and infrastructure-as-code have different lifecycles and blast radii. You redeploy a bundle many times a day; you create a catalog or a service principal once, carefully, with a separate approval path. Mixing them means a routine `bundle deploy` could churn a catalog or rotate an identity — exactly what you don't want. Keep code in the bundle, infra in Terraform, and let the bundle *reference* the infra by name.
