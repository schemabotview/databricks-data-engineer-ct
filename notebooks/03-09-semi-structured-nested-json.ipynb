## Semi-structured & nested data — JSON into Delta

The card-transactions feed is nested JSON — each transaction has nested `card`, `merchant`, and `auth` sub-objects. Two options:

**Land it nested, flatten in silver** (the pattern the exam rewards). Auto Loader infers a nested `STRUCT` / `ARRAY<STRUCT>` schema; the bronze table keeps it nested, and silver flattens with dot-access, `explode()`, and `from_json()`. Keeps the original shape and replay-ability.

**Land it flat at the source** — configure the upstream to emit flattened JSON. Cleaner downstream, but you lose the original shape.

Three primitives to memorise:

- **Dot access** — `df.select("merchant.name", "merchant.category")` on `STRUCT` fields.
- **`explode(arr)`** — one row per array element (unpacking line-items / events).
- **`from_json(str, schema)`** — parse a stringified JSON column into a struct; the way you turn `_rescued_data` into typed fields.

For a JSON file that is one object spread across lines (not newline-delimited), use `multiLine = true` — supported by Auto Loader as a format option on `cloudFiles.format = json`.
