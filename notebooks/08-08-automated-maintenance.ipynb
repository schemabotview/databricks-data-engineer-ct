## Automated maintenance — Liquid Clustering & Predictive Optimization

Two features that keep Delta tables healthy without a hand-written maintenance job. Both were introduced in module 02; here's the operational angle the exam tests.

**Liquid Clustering — evolvable data layout.** Pre-Liquid, you picked a partition column up front; getting it wrong meant a full TB rewrite. Liquid Clustering lets you declare cluster keys that Delta re-balances incrementally — and **change them later without rewriting the world**:

```sql
CREATE TABLE silver.card_transactions (...) CLUSTER BY (customer_id, transaction_at);
-- months later, query patterns shifted:
ALTER TABLE silver.card_transactions CLUSTER BY (merchant_category, transaction_at);
```

Operationally it means: no small-files-per-partition problem, no skew when one partition is 100× larger (clusters re-balance), and evolvable keys. **It's the recommended default for new Delta tables.**

**Predictive Optimization (PO) — Databricks runs maintenance for you.** It runs `OPTIMIZE` (compaction + cluster rebalancing) and `VACUUM` (storage reclamation) at the right cadence, informed by actual access patterns, billed at a serverless rate.

```sql
ALTER CATALOG fintech_prod ENABLE PREDICTIVE OPTIMIZATION;
```

Two constraints: **Unity Catalog managed tables only** (external tables don't get it — one more reason to prefer managed), and specific regions/DBRs (the exam treats it as universal). **Exam pattern:** *"keep gold tables compact without writing a maintenance job"* → **enable Predictive Optimization**.
