## The problem CI/CD solves on Databricks

Before bundles, promoting a pipeline from dev to prod meant one of two bad options.

**Option A — point and click.** Engineers cloned notebooks across workspaces by hand, copy-pasted job JSON, and updated paths manually. Drift was inevitable: prod ran a job spec slightly different from dev, and the two diverged over months.

**Option B — REST API scripts.** Teams wrote bash that called the Jobs API and Workspace API to upload notebooks and create jobs. Workable, but every team reinvented the same plumbing differently — there was no canonical way to describe *"this pipeline plus this job plus these notebooks, deployed as one unit."*

**Automation Bundles** solve this with a single declarative file — **`databricks.yml`** — that describes every resource in your project, and a **CLI** that validates, deploys, and runs that bundle against any target workspace. The same source tree promotes from **dev → test → prod** by changing one CLI flag.

The worked example for this module is the bank's `fintech-platform` Git repo, which holds the ingestion / silver / gold notebooks, the declarative pipeline (module 05), the `fintech_nightly_ingest` job (module 06), and one `databricks.yml` that promotes the whole bundle across three workspaces.

**The exam's framing:** *"a team needs a modular, version-controlled way to deploy and orchestrate ETL with CI/CD and repeatability."* The answer is **Automation Bundles in Git, deployed with `bundle deploy`** — never hand-cloning notebooks or snapshotting the UI.
