## Performance tuning knobs — the four the exam asks about

You won't deep-tune here (module 08 owns that), but the exam's Section 3 calls out four config keys by name. Know what each controls.

| Config | Controls | Default | Change when |
|---|---|---|---|
| `spark.sql.shuffle.partitions` | Partitions after a shuffle (joins, aggregations) | 200 | Big jobs: raise (1000+) to shrink per-partition data and avoid OOM. Small jobs: drop (8) to avoid empty tasks. AQE coalesces this dynamically. |
| `spark.default.parallelism` | Default partition count for RDD ops | `# cores` | Rarely tuned on DataFrame work — mostly an RDD knob. |
| `spark.executor.memory` / `spark.driver.memory` | Heap per executor / driver | cluster-dependent | Executor OOM → raise executor memory; collect-to-driver crash → raise driver memory. |
| `spark.sql.autoBroadcastJoinThreshold` | Max size Spark will auto-broadcast | 10 MB | Raise to 100 MB+ for larger dimension tables; set `-1` to disable auto-broadcast. |

**The way the exam phrases it:** *"a job is shuffling 200 GB across 200 partitions and tasks are spilling to disk — what should you change?"* → raise `spark.sql.shuffle.partitions` so each partition lands in the **100–200 MB** range. Too few partitions → spill and OOM; too many → hundreds of tiny empty tasks. AQE (module 08) coalesces shuffle partitions automatically when enabled, which is why modern clusters lean on it instead of hand-setting 200.
