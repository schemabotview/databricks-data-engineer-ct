## Lakeflow Jobs run history — trend, not snapshot

The **run history view** lists every run of a job with its state, start time, duration, and a **sparkline of recent durations**. Its single most valuable signal: **trend against the historical baseline** — not what one run did, but how this run compares to the last hundred.

Three shapes to recognise:

- **Gradual ramp** — duration creeping up over weeks. Usually **data growth** outpacing cluster capacity, or a slowly-growing fact table making a join progressively skewed.
- **Step change** — one day where the curve jumps and stays up. Usually a **code or config change**, a new upstream source, or a runtime upgrade.
- **Spiky** — most runs at 10 min, an occasional run at 90. Usually **concurrency / resource contention**, or an occasional bad-data day.

Each shape points at a different cause, so the shape *is* the first diagnosis: a ramp says "capacity / growth," a step says "something changed on date X," a spike says "contention."

**The exam's framing:** *"a job's median duration tripled overnight — where do you look first?"* → the **run history view** for the trend, then the **task graph** of a slow run. A single slow run tells you little; the trend tells you whether this is new, gradual, or a one-off.
