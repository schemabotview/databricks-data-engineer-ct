## Databricks Runtime & Photon

Every cluster boots a **Databricks Runtime (DBR)** — a pre-built image of a tuned Apache Spark plus optimized libraries, Delta Lake, and the operating system. Runtimes are **versioned** (for example, DBR 15.x), and each release pins specific Spark and library versions so your environment is reproducible. For production, prefer an **LTS (long-term support)** version for stability.

There are a few variants. The **standard** runtime covers data engineering; the **ML runtime** adds machine-learning libraries and GPU support for training workloads.

**Photon** is Databricks' **vectorized query engine**, written in C++. It's a drop-in replacement for Spark's execution engine that accelerates **SQL and DataFrame** workloads — the same query runs faster and, because it finishes sooner, usually cheaper. It's **transparent to your code**: you don't rewrite anything; you enable Photon on a supported cluster or SQL warehouse and eligible operations run on the native engine.

For the exam: **DBR is the runtime image** (Spark + Delta + libraries), pick **LTS** for production and the **ML runtime** for training; **Photon is the C++ vectorized engine** that speeds up SQL and DataFrame operations with no code changes.
