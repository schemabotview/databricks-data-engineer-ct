## SQL warehouses & serverless compute

Not every workload is a Spark notebook. **SQL warehouses** are compute tuned specifically for **SQL and BI**: they power Databricks SQL, dashboards, the SQL editor, and external BI tools like Power BI and Tableau. A warehouse absorbs many concurrent queries by **autoscaling** — adding more clusters as query load rises, removing them as it falls.

Warehouses come in three types: **Classic** and **Pro** run in *your* cloud account (Pro adds advanced features), while **Serverless** runs in Databricks' account and starts in seconds with the best price/performance for SQL.

That last one points at a broader idea. **Serverless** is a *modality*, not a single product: Databricks manages the compute in its own account, it starts in seconds, autoscales, and you pay only for what you use — with **no idle cost**. Serverless is available for **SQL warehouses, jobs, notebooks, and pipelines**.

The trade-off is the same one from the architecture section: with serverless you give up some network control (it runs in Databricks' account), and in return you get instant start and zero cluster management. For the exam: **SQL warehouse = compute for SQL/BI; serverless = the fast-start, Databricks-managed option across SQL, jobs, notebooks, and pipelines.**
