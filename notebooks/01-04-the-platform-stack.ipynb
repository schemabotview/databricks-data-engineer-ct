## The platform stack at a glance

Zoom out and the platform is a stack of four layers, each depending on the one below it.

**Cloud object storage** is the base — S3, ADLS, or GCS: the cheap, durable place your bytes actually sit.

**Delta Lake** sits on top of storage. It's the open table format that turns loose files into transactional tables with ACID guarantees, schema, and time travel.

**Unity Catalog** sits above Delta. It's the governance layer that names, secures, and audits every table, file, and model through one three-level namespace.

**Engines and workloads** run on top — the Spark and Photon engines, Databricks SQL, and Lakeflow ingestion, pipelines, and jobs — all reading *through* Unity Catalog to the same Delta tables.

You'll go deep on the two foundation layers, Delta Lake and Unity Catalog, in the next module. For now the point is the shape: cheap open storage at the base, transactions and governance in the middle, and every engine on top sharing one copy of the data.
