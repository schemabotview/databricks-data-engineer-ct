## Privileges — what each one buys you

The list the exam expects you to recognise on sight:

| Privilege | Applies to | Lets the grantee… |
|---|---|---|
| `USE CATALOG` | catalog | see schemas inside |
| `USE SCHEMA` | schema | see objects inside |
| `SELECT` | table / view / MV | read rows |
| `MODIFY` | table / view | `INSERT` / `UPDATE` / `DELETE` / `MERGE`; `ALTER` for schema |
| `CREATE`, `CREATE TABLE`, `CREATE SCHEMA` | catalog / schema | create objects inside |
| `EXECUTE` | function | call the function |
| `READ VOLUME` / `WRITE VOLUME` | volume | read / write files |
| `BROWSE` | catalog / schema | see metadata without seeing data |
| `APPLY TAG` | object | tag for governance (used by ABAC) |
| `ALL PRIVILEGES` | any | every privilege at that level (use sparingly) |

**Ownership is separate from privileges.** The **owner** of an object can do anything to it, *including changing its grants*. There is exactly **one owner** per object; `ALTER ... OWNER TO <group>` transfers it.

**Two distinctions the exam tests:** `SELECT` (read) vs. `MODIFY` (write — the one privilege that covers `INSERT`/`UPDATE`/`DELETE`/`MERGE`), and `BROWSE` (see that an object *exists*, and its metadata) vs. `SELECT` (actually read its rows). Prefer owning objects with **groups**, not individuals, so ownership survives someone leaving.
