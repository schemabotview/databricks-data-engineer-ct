## The five gold object types

The exam expects you to pick the right one for each scenario. Memorise the shape and the "right when":

| Object | What it is | Stored? | Right when… |
|---|---|---|---|
| **Table** | Plain Delta you write with batch jobs | ✅ | You need full control over how & when it's written |
| **View** | Logical query, evaluated on read | ❌ | Light wrapping, security barriers, hiding PII |
| **Materialized View (MV)** | Precomputed result, stored, auto-refreshed | ✅ | Heavy aggregation BI reads often — `customer_360` |
| **Streaming Table** | Append-only, fed by a Structured Streaming query | ✅ | Records arrive continuously — `daily_card_volume` |
| **MV / ST in a Declarative Pipeline** | Same shapes, defined declaratively with expectations + lineage | ✅ | You want the framework's quality + lineage built in |

**Plain views are the security-barrier tool.** A view stores no data — it's a saved `SELECT`, evaluated every read. Hide PII by selecting only safe columns, then grant on the view:

```sql
CREATE VIEW fintech_dev.gold.customers_public AS
SELECT customer_id, city, country, created_at
FROM   fintech_dev.silver.customers;   -- no email / phone / DOB exposed
```

Temporary views are session-scoped: `df.createOrReplaceTempView("t")`, or `createOrReplaceGlobalTempView` for cross-session under `global_temp.`.

**The trade-off the exam tests:** a view **recomputes on every read**. For a heavy aggregation hit by a thousand dashboard refreshes a day, a **materialized view** wins; a plain view would re-run the aggregation every single time.
