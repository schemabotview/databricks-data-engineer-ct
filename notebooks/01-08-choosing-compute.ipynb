## Choosing compute — the four-scenario cheat sheet

The exam is scenario-driven: it hands you a workload and asks which compute fits. Here's the cheat sheet.

- **Interactive development / ad-hoc exploration in notebooks** → **all-purpose cluster**. It stays up, multiple people can share it, and you can rerun cells as you iterate.
- **Scheduled, automated production ETL** → **job cluster**. It spins up per run, is isolated, tears down when done, and bills at the cheaper job rate.
- **BI dashboards and ad-hoc SQL / connecting BI tools** → **SQL warehouse** (prefer **serverless** for the fastest start and best price/performance).
- **Bursty or unpredictable work, or "I don't want to manage infrastructure"** → **serverless compute**. Seconds to start, autoscales, no idle cost.

One more that comes up: **ML training** runs on an **all-purpose or job cluster using the ML runtime** (which bundles ML libraries and can use GPUs).

The throughline: match the *shape* of the work to the compute. Interactive and shared → all-purpose. Scheduled and isolated → job. SQL and BI → warehouse. Spiky and hands-off → serverless.
