## What the Data Intelligence Platform is

The Databricks **Data Intelligence Platform** is the productized lakehouse: one place where data engineering, SQL analytics, streaming, data science, and machine learning all run against a **single governed copy** of data. It's built in three layers.

**1. An open lakehouse foundation.** Your data lives as Delta Lake tables on *your own* cloud object storage, in open formats — so no vendor locks the bytes away and any engine can read them.

**2. Unified governance.** Unity Catalog governs every table, file, model, and dashboard through one permission model that spans all workspaces and clouds — not a separate ACL system per tool.

**3. A data intelligence engine.** This is what puts *intelligence* in the name: a layer that uses generative AI to understand the semantics of your data, powering natural-language search, automatic optimization, and smart defaults so more people can use the platform without deep tuning.

The payoff is one platform, one copy of data, every workload and persona — instead of stitching a warehouse, a lake, a streaming system, and an ML platform together and copying data between them. (Databricks recently renamed this from the *Lakehouse Platform* to the *Data Intelligence Platform* to reflect that AI layer.)
