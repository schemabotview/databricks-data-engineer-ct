## Unity Catalog & the three-level namespace

Delta Lake makes the *files* trustworthy. **Unity Catalog (UC)** makes the *tables, views, volumes, functions, and models* discoverable and governed — one catalog across SQL, Python, ML, and dashboards. Three things the legacy hive metastore couldn't do: a **unified permission model** (GRANT/REVOKE on principals at any level), **lineage and audit out of the box**, and **cross-workspace, cross-cloud** namespacing (one metastore per region).

The namespace is **three levels** — every object has a three-part name:

```text
metastore  (one per region)
└── catalog: fintech_dev              catalog.schema.object
     ├── schema: bronze
     │    ├── table:  card_transactions
     │    ├── view:   recent_txns
     │    ├── volume: landing_zone       (files, not rows)
     │    └── function / model / materialized view
     ├── schema: silver
     └── schema: gold
```

Six **securable** object types live at the schema level: tables, views, volumes, functions, models, and materialized views. Permissions **flow down**: `USE CATALOG fintech_dev` + `SELECT` on `fintech_dev.gold` lets a principal read all of gold without touching bronze or silver. (Full GRANT semantics in module 09.)

**Legacy callout — `hive_metastore`.** Every pre-UC workspace still has a special `hive_metastore` catalog holding legacy two-part `schema.table` names. It's **per-workspace, not regional** — no cross-workspace governance — and new tables should **never** land there. When the exam contrasts `hive_metastore.default.foo` with `fintech_dev.silver.foo`, the three-part UC name is always right.
