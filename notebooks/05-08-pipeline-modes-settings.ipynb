## Pipeline modes, settings & the development workflow

A declarative pipeline is configured once and re-run many times. Four knobs the exam tests:

**Triggered vs. Continuous** — *how the pipeline runs.*

- **Triggered** — runs once against all available data, then stops. Best for scheduled batch. Lower cost.
- **Continuous** — runs forever, processing new data as it arrives. Best for real-time. Higher cost.

**Development vs. Production** — *how the cluster behaves.*

- **Development** — cluster stays alive between runs (fast iteration); errors don't tear the cluster down.
- **Production** — cluster terminates between runs (cheaper); errors fail the run.

**Channel** — *which runtime.*

- **Current** — the latest GA runtime. **Preview** — early-access features.

**Target catalog & schema** — every pipeline lands its datasets into a UC catalog + schema you choose. Without a target, the datasets are visible only to the pipeline itself.

**The exam's angle:** don't confuse the two axes. *Triggered vs. continuous* is about **data cadence**; *development vs. production* is about **cluster lifecycle and error handling**. A pipeline can be triggered-and-production (scheduled nightly batch) or continuous-and-development (iterating on a streaming pipeline).
