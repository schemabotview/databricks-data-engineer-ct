## What is a Lakeflow Job?

Three nested concepts — get the vocabulary right and everything else in this module falls into place.

- **Job** — the unit of orchestration: a named, version-controlled definition of *what to run and when*. The bank runs one nightly job per business unit.
- **Task** — one step in a job (a notebook, SQL, pipeline, dashboard, …). A job is one task, or many tasks wired into a DAG.
- **Run** — a single execution. *"The job had 50 runs last month; 49 succeeded."* Every run has a state: pending / running / succeeded / failed / cancelled / skipped.

**Rename watch.** *Lakeflow Jobs* is the current name; *Databricks Workflows* and *Databricks Jobs* are the legacy names — the exam may use any of the three for the same thing.

**Single-task vs. multi-task.** A one-task job is fine for trivial work, but the exam mostly tests **multi-task jobs** — a DAG of dependent tasks — because that's what real ETL looks like: ingest in parallel, build silver, refresh gold, fire a dashboard, notify.

The worked example for this module is the bank's nightly **`fintech_nightly_ingest`** job: it ingests the daily card / account / loan drops, runs silver transformations, refreshes `customer_360`, fires the BI dashboard, and emails ops on failure.
