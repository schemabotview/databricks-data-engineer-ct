## `GRANT` / `REVOKE` / `DENY` — on principals

Three SQL statements, three roles in the system.

**`GRANT`** adds a privilege — walk the hierarchy top-down:

```sql
GRANT USE CATALOG ON CATALOG fintech_dev            TO `analysts`;
GRANT USE SCHEMA  ON SCHEMA  fintech_dev.gold       TO `analysts`;
GRANT SELECT      ON TABLE   fintech_dev.gold.customer_360 TO `analysts`;
```

**`REVOKE`** removes a privilege that was explicitly granted:

```sql
REVOKE SELECT ON TABLE fintech_dev.silver.customers FROM `analysts`;
```

**`DENY`** explicitly *blocks* a privilege, even if another grant would allow it:

```sql
DENY SELECT ON TABLE fintech_dev.silver.customers TO `analysts`;
```

**The precedence rule the exam tests: `DENY` beats `GRANT`.** If you granted `SELECT ON CATALOG fintech_dev` but denied `SELECT ON TABLE fintech_dev.silver.customers`, the deny wins for that one table — the analyst reads everything in the catalog *except* that table.

**Three principal types:** **users** (human accounts), **groups** (preferred for production — manage membership in your IdP, grant to the group name in UC), and **service principals** (non-human identities for jobs / CI-CD, module 07).

**Best practice the exam rewards: grant to groups, never to individual users.** Group grants survive joiners and leavers; per-user grants rot into an unauditable mess.
