## The decision sheet — picking the right gold object

The whole module on one page. Read the scenario, pick the object:

| Scenario | Pick |
|---|---|
| Heavy aggregation BI dashboards refresh many times an hour | **Materialized view** |
| Continuous fact ingestion — the gold table should reflect a live stream | **Streaming table** |
| Light wrapper / hide PII columns / encode a common filter | **View** |
| Full declarative pipeline with quality gates, lineage, CDC | **Lakeflow Spark Declarative Pipelines** |
| SCD type 2 history on a dimension fed by CDC | **`APPLY CHANGES INTO ... STORED AS SCD TYPE 2`** |
| Drop rows that fail a quality rule, keep the pipeline going | **`expect or drop`** |
| Stop the pipeline on a critical violation | **`expect or fail`** |
| Reject bad rows at write time on a plain Delta table | **`CHECK` constraint** |
| Custom batch logic you fully control end to end | **Plain Delta table + Lakeflow Jobs task** |

**Three tells that unlock most questions:**

- *"reads it a lot, refresh on a schedule"* → **materialized view**.
- *"arrives continuously"* → **streaming table**.
- *"keep full history of changes"* → **`APPLY CHANGES ... SCD TYPE 2`**.

Everything else is a variation on *view* (cheap wrapper), *`CHECK`* (write-time plain-Delta guard), or *expectations* (pipeline-quality by severity).
