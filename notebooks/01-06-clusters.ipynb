## Clusters — all-purpose vs. job

A **cluster** is a set of cloud VMs running Spark together: one **driver** that coordinates and one or more **workers** that do the parallel work. Databricks gives you two kinds, and the exam leans on the difference.

**All-purpose clusters** are *interactive*. You create one by hand, attach notebooks to it, and it stays up while you explore — multiple users can share it. This is the compute for **development, ad-hoc analysis, and debugging**. Because it lingers and serves interactive work, it bills at a **higher DBU rate**.

**Job clusters** are *ephemeral*. A Lakeflow Job creates one automatically when a run starts and tears it down the moment the run finishes. Each run gets a fresh, single-purpose cluster — so runs are **isolated** from each other, and the **DBU rate is lower**. This is the compute for **scheduled, automated production ETL**.

The rule of thumb: **dev and exploration → all-purpose; scheduled production → job clusters.** Running a production job on an all-purpose cluster is a classic anti-pattern — it costs more and loses the per-run isolation a job cluster gives you.
