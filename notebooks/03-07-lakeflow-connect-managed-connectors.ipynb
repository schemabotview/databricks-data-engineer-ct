## Lakeflow Connect & managed connectors

`COPY INTO` and Auto Loader solve *files arriving in object storage*. But the bank also has data in **operational systems** — PostgreSQL for loans, Salesforce for cases, NetSuite for finance. **Lakeflow Connect** is the managed ingestion family for those, in three peer tiers:

1. **Managed connectors** — fully Databricks-operated pipelines for SaaS and database sources. Give credentials + a target catalog; Databricks runs the connector, does CDC, schema evolution, retries, and monitoring.
2. **Standard connectors** — the open building blocks: Auto Loader for files, Structured Streaming for Kafka / Kinesis / Event Hubs / Pub/Sub. You write the code; Databricks provides the runtime.
3. **Partner connectors** — Fivetran, Airbyte, dbt, etc., via Partner Connect.

All three land into UC-governed Delta. A **managed connector** is the low-effort default where the source is supported:

- **SaaS** — Salesforce, Workday, ServiceNow, NetSuite, SharePoint.
- **Databases (with CDC)** — SQL Server, PostgreSQL, MySQL, Oracle.
- **Events** — Kafka via Confluent.

The bank enables the **Salesforce managed connector**, picks objects (`Case`, `Contact`, `Account`), points it at `fintech_dev.bronze.salesforce`, and Databricks handles incremental CDC sync — no custom REST poller. The exam likes managed connectors for lowest operational burden, native CDC, built-in monitoring, and pre-governed UC tables.
