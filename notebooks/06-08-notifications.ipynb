## Notifications & operational hooks

How a job talks to humans and systems when things happen. Four hooks.

**Email notifications** — on start, success, failure, duration warning, or streaming backlog. Configured at the job level.

**Webhooks / system destinations** — Slack, Microsoft Teams, PagerDuty, or a generic HTTP endpoint. **The right answer when the question mentions paging on-call.**

**Run-as identity** — a job runs *as* a user or a service principal, and that identity governs which Unity Catalog permissions apply. **Production jobs should run as a service principal**, not a human's account — so the job keeps working when that person leaves or loses access.

**Always-run-on-finish task** — wire a final task with `run_if = ALL_DONE`; it fires whether upstream succeeded or failed. The bank uses this to push a summary status to Slack after every nightly run.

**Two exam tells:**

- *"page the on-call team on failure"* → a **webhook / system destination** (Slack / PagerDuty), not just email.
- *"a production job should not depend on one engineer's credentials"* → **run as a service principal**.

Notice the two ways to guarantee a notification fires on failure: a job-level `on_failure` email/webhook, or a final `ALL_DONE` task. The `ALL_DONE` task also lets you *do work* (write a summary, update a status table), not just send a message.
