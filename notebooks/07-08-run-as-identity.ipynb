## Run-as identity — service principals in non-dev

Bundles execute under an identity, and there are two distinct questions:

- **Who deploys the bundle?** The principal that runs `databricks bundle deploy`. In CI this is a **service principal** authenticated via OAuth machine-to-machine.
- **Who runs the deployed resources?** Set by `run_as` on each target.

**Best practice the exam expects:**

- **dev** → `run_as: user` — the developer's own identity.
- **test / prod** → `run_as: { service_principal_name: <name> }` — so production jobs never depend on a human's account.

**Why a service principal in prod:** if a job runs as a person, it breaks the day that person leaves the team, changes roles, or has their token rotated. A service principal is a stable, non-human identity owned by the platform — the correct answer whenever a question says *"a production job should not depend on any one engineer's credentials."*

**The permission boundary** — the service principal needs, on what it touches:

- **`USE CATALOG`** + **`USE SCHEMA`** on the catalogs / schemas.
- **`SELECT`** / **`MODIFY`** on tables.
- **`CAN_MANAGE_RUN`** on the bundle's jobs / pipelines.
- **Workspace folder permissions** for the deploy path.

Module 09 covers `GRANT` / `REVOKE` on these in detail. The takeaway here: prod is a service principal, and it holds exactly the grants it needs — no more.
