## The diagnostic flow — where to look first

The bank's `silver_build` task used to finish in 12 minutes; last week it started taking 38. When a job misbehaves, follow this sequence — the exam asks variants of it directly.

```text
① Lakeflow Jobs — run history view
     Is THIS run slow, or have all recent runs been slow?  → a trend line
② Lakeflow Jobs — task graph of one slow run
     Which TASK is slow (and is it slow because an upstream is)?  → "silver_build"
③ Spark UI — Stages tab
     Which STAGE is slow?  → "the join stage: 30 min"
④ Stage detail — task summary metrics
     Min/Median/Max shuffle read, spill, GC  → "Max shuffle read 5 GB vs median 400 MB"
⑤ Fix — match the pattern to the remedy (the rest of this module)
```

**Work top-down, coarse to fine:** job → task → stage → task-metrics → fix. Never start by bumping cluster size — you can't fix what you haven't localised.

**The exam loves step ④.** *"Most tasks finish quickly but one takes much longer; Max shuffle read is 10× the median."* That sentence is **skew**, and the recommended answer is **AQE skew-join**. Learn to read the min/median/max summary — it's where the diagnosis actually lives, and where most of this module's fixes get chosen.
