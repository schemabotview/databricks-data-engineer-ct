## Liquid Clustering — the modern replacement for partitioning + `ZORDER`

Classic Delta layouts force a one-time decision: partition by `transaction_date`? By `customer_id`? Get it wrong and you're stuck — re-partitioning a terabyte-scale table is a full rewrite. Over-partitioning also creates the small-files-per-partition problem, and skew when one partition is 100x larger than another.

**Liquid Clustering** replaces both partitioning *and* `ZORDER` with a single declared property. You name the cluster columns, and Delta organises data into clusters it can incrementally re-balance — and you can **change the keys later without rewriting the world**.

```sql
CREATE TABLE silver.card_transactions (...)
  CLUSTER BY (customer_id, transaction_at);

-- Change keys later — Delta reorganises incrementally
ALTER TABLE silver.card_transactions
  CLUSTER BY (merchant_category, transaction_at);
```

**What you give up:** the explicit directory-style partition folders. **What you gain:** no small-files-per-partition problem, no partition skew, and clustering that evolves as query patterns change.

**For the exam:** Liquid Clustering is the **recommended layout for new Delta tables**.
