## The Databricks CLI — `bundle validate / deploy / run / destroy`

Four subcommands cover the entire bundle workflow. Know what each does.

- **`databricks bundle validate -t <target>`** — parses the manifest, resolves variables for the target, checks all references, and emits the **fully-resolved JSON** that would be sent to the API. Run it first; run it in CI on every PR. It's the dry-run / lint step.
- **`databricks bundle deploy -t <target>`** — uploads notebooks and source files, then creates/updates every resource so the workspace **matches the manifest**. **Idempotent** — re-deploying an unchanged bundle is a no-op.
- **`databricks bundle run <resource_key> -t <target>`** — kicks off a one-off run of a specific job or pipeline and tails its output. The CLI equivalent of clicking *Run now*.
- **`databricks bundle destroy -t <target>`** — removes all resources the bundle owns from the target. Use carefully; use it in CI to tear down ephemeral test environments.

Typical local loop:

```bash
databricks bundle validate -t dev
databricks bundle deploy   -t dev
databricks bundle run      fintech_nightly_ingest -t dev
```

**Auth** — the CLI authenticates via an **OAuth user login** (`databricks auth login`) for local dev; an **OAuth machine-to-machine** service-principal token in CI (`DATABRICKS_HOST` + `DATABRICKS_CLIENT_ID` / `_SECRET`); or a **PAT** (older, still works, less preferred).

**The exam's command question:** the one that *parses the manifest, resolves variables, checks references, and emits the JSON it would send* is **`validate`** — not `deploy`, `init`, or `run`.
