## Common failures — cluster startup & library conflicts

Two failure classes the exam tests by scenario. When a cluster fails to start, the **Event Log** tab is the source of truth.

**Cluster startup — four modes:**

- **Cloud capacity / quota** — `InstanceLimitExceeded`, `InsufficientCapacity`. The instance type isn't available in that region/AZ, or your quota is exhausted. → change instance type, change AZ, request a quota increase, or **switch to serverless**.
- **IAM / permissions** — `NotAuthorizedToAssumeRole`, `403 from STS`. The cluster's instance profile / SP can't reach a resource. → fix the role's trust + attached policies.
- **Init script failing** — `INIT_SCRIPT_FAILURE` (e.g. a `pip install` hitting a network policy). → read the init-script log, fix, restart. *(Init scripts don't run on serverless — one more reason it dodges this class.)*
- **Library install failing** — a bad PyPI/Maven library leaves the cluster in `ERROR`. → find it in the *Libraries* tab, remove or pin.

**Library conflicts** — two packages needing incompatible versions of the same dependency; symptoms are a deep `ImportError`, or "worked yesterday, broke after the DBR upgrade." Resolution order:

- **Notebook-scoped `%pip install`** — each notebook gets its own env; the fix when only one notebook needs the dependency.
- **Pin versions** in cluster libraries (`mypkg==2.4.1`, not `mypkg`).
- **Use the DBR ML runtime** for ML packages (pre-resolved), or **serverless** (Databricks resolves the stack; you just `%pip install`).

**Exam tells:** `InstanceLimitExceeded` → a cloud **quota/capacity** issue (not a memory setting). Two notebooks colliding on a shared cluster → **notebook-scoped `%pip`** (or serverless).
