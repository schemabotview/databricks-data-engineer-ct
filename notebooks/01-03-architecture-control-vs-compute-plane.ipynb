## Architecture — control plane vs. compute plane

Every Databricks deployment splits into two halves, and knowing which is which explains where your code runs and where your data lives.

The **control plane** is managed by Databricks, in Databricks' own cloud account. It holds the web UI, the notebook and SQL editors, the job scheduler, the cluster manager, query history, and the Unity Catalog metadata. Think of it as the brains and the management services — it stores *metadata and results*, but not your bulk data.

The **compute plane** is where your data is actually processed — the clusters and SQL warehouses running Spark. It comes in two flavors:

- **Classic compute** runs in *your* cloud account, inside your own network (VPC / VNet). Your data never leaves your account, which suits stricter isolation requirements.
- **Serverless compute** runs in Databricks' account and is available almost instantly, with the network managed for you — you trade some network control for speed and zero cluster management.

In both cases your **data at rest** stays in *your own* cloud object storage — it never sits in the control plane. So the mental model for the exam is: notebooks, metadata, and results in the control plane; data processing in the compute plane; data at rest in your storage.
